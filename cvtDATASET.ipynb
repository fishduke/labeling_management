```python

import cv2
import os
import glob
import xml.etree.ElementTree as elemTree

file = open(r"./yolov4_prune/data/lpr1val.txt", "r")
strings = file.readlines()
cvt_string = []
for i in range(len(strings)):
    num = len(strings[i])-1
    data_path = strings[i][0:strings[i][0:num].rfind('/')+1]
    jpg_name = strings[i][0:strings[i].rfind('.')] + ".jpg"
    sub = strings[i][0:num].rfind('/')+1
    file_name = strings[i][sub:]
    lable_name = file_name[0:file_name.rfind('.')] + ".xml"
    lable_path = data_path + lable_name

    xml_file = elemTree.parse(lable_path)
    main_string = ""
    for xml_object in xml_file.findall('object'):
        obj_name = ""
        for obj in xml_object:
            element = obj.get(obj.tag)
            if(obj.tag == 'name'):
                obj_name = obj.text

        if(len(main_string)>2):
            width = ""
            height = ""
        else :
            img = cv2.imread(strings[i][0:strings[i].rfind('.')] + ".jpg")
            width = str(img.shape[1])
            height = str(img.shape[0])
            main_base = " " + width + " " + height
        
        object_xml = " " + obj_name + " " + obj[0].text + " " + obj[1].text + " " + obj[2].text + " " + obj[3].text
        main_string = main_string + object_xml
    if(len(main_string)>1):
        main_string = jpg_name + main_base + main_string + '\n'
        cvt_string.append(main_string)    
    
    cvt_file = open(r"./yolov4_prune/data/lpr1cls2.txt", "w")
    cvt_file.writelines(cvt_string)

file.close()
cvt_file.close()
```

# ### if label data is out of frame, make it right ###

```python


import cv2

def draw_rectangle(image, x_center, y_center, width, height):
    # 이미지의 너비와 높이
    img_height, img_width = image.shape[:2]

    # 박스의 좌상단, 우하단 좌표 계산
    x1 = int((x_center - width / 2) * img_width)
    y1 = int((y_center - height / 2) * img_height)
    x2 = int((x_center + width / 2) * img_width)
    y2 = int((y_center + height / 2) * img_height)

    # 이미지에 사각형 그리기
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    return image

def adjust_label_to_frame(label, frame_width, frame_height):
    # label 형식: <object-class> <x_center> <y_center> <width> <height>
    object_class, x_center, y_center, width, height = label.split()

    # 박스 좌표 계산
    x1 = float(x_center) - float(width) / 2
    y1 = float(y_center) - float(height) / 2
    x2 = float(x_center) + float(width) / 2
    y2 = float(y_center) + float(height) / 2

    # 프레임 내에 있는지 확인
    if x1 < 0:  # 왼쪽을 벗어난 경우
        width = float(width) + x1  # 너비 조절
        x1 = 0  # 좌표 조정
    if y1 < 0:  # 위를 벗어난 경우
        height = float(height) + y1  # 높이 조절
        y1 = 0  # 좌표 조정
    if x2 > 1:  # 오른쪽을 벗어난 경우
        width = float(width) - (x2 - 1)  # 너비 조절
        x2 = 1  # 좌표 조정
    if y2 > 1:  # 아래를 벗어난 경우
        height = float(height) - (y2 - 1)  # 높이 조절
        y2 = 1  # 좌표 조정

    # 수정된 좌표로 새로운 라벨링 값 생성
    new_width = round(x2 - x1, 6)
    new_height = round(y2 - y1, 6)
    new_x_center = round(x1 + new_width / 2, 7)
    new_y_center = round(y1 + new_height / 2, 7)

    adjusted_label = f"{object_class} {new_x_center} {new_y_center} {new_width} {new_height}"
    return adjusted_label

def make_label_equal(path):
    images = glob.glob(path + '*.jpg')
    labels = glob.glob(path + '*.txt')

    count = 0
    cvt_class = ''
    for image in images:
        label = image[:image.rfind('\\')+1] + image[image.rfind('\\')+1:len(image)-4] + ".txt"
        
        try:
            file = open(label, "r")
            if(0 < len(file.readlines())):
                file = open(label, "r")
                if(file.readline()[:2]=='-1'):
                    file = open(label, "r")
                    cvt_class = '0 ' + file.readline()[3:]
                    file = open(label, "w")
                    file.write(cvt_class)
                    file.close()

            count += 1
            
        except:
            try:
                file = open(label, "r")
                if(0 == len(file.readlines())):
                    file.close()
            except:
                file = open(label, "w")
                file.close()

import glob

path = ''
make_label_equal(path)

images = glob.glob(path + '*.jpg')
labels = glob.glob(path + '*.txt')

cvt_class = ''
nonlabel = 0
labeled = 0
count = 0
for image in images:
    count += 1
    label = image[:image.rfind('\\')+1] + image[image.rfind('\\')+1:len(image)-4] + ".txt"
    img = cv2.imread(image)
    
    try:
        file = open(label, "r")
        lines = file.readlines()
        file.close()
        if(0 < len(lines)):
            file = open(label, "w")
            for i in range(len(lines)):
                object_class, x_center, y_center, width, height = lines[i].split()
                adjusted = adjust_label_to_frame(lines[i], float(width), float(height))
                if(i != len(lines)-1):
                    adjusted = adjusted + "\n"
                file.write(adjusted)
        labeled += 1
        file.close()
            
    except:
        nonlabel += 1
        
    msg = "\rprocessed : %.0f%%" % (count/len(images)*100.0)
    print(msg,end='')

print("\nlabeled :", labeled)
print("nonlabel :", nonlabel)
```
