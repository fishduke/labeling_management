## Making train & valid & data & names ##

```python
#for shuffle
import random

original_path = "/data2/fire/2023-10-31/fire.txt"
out_path = "/data2/fire/2023-10-31/fire_split.txt"
darknet_path = "/home/fishduke/Desktop/darknet/data/"
f_path = darknet_path + "fire_train.txt"
f2_path = darknet_path + "fire_valid.txt"
data_path = darknet_path +"fire.data"
names_path = "fire.names"

classes = ['fire','smoke']
do_split = 1

txt = open(original_path,'r')
tmp = []

while True :
    line = txt.readline()
    if not line:
        break
    tmp.append(line)
txt.close()

if(do_split):
    random.shuffle(tmp)

f = open(out_path,'w')        
for i in tmp :  
    f.write(i)
f.close()

#for create
count = 0
length = 5196 #total line

txt = open(out_path,'r')

i = 0
f = open(f_path,'w')
f2 = open(f2_path,'w')

while True :
    if i == 0 :
        line = txt.readline()
        if not line :
            break
        count +=1
        if count < int(length/10)*1 :
            f2.write(line)
        else :
            f.write(line)

txt.close()
f.close()
f2.close()

data = open(data_path, 'w')
data.write('classes= '+str(len(classes))+'\n')
data.write('train = ' + f_path+'\n')
data.write('valid = ' + f2_path+'\n')
data.write('names = ' + darknet_path + names_path+'\n')
data.write('backup = ./backup'+'\n')
data.write('eval=coco')
data.close()

names_path = darknet_path + names_path

names = open(names_path, 'w')
count = 0
for name in classes:
    names.write(name)
    count += 1
    if(len(classes) > count):
        names.write('\n')
names.close()
```
